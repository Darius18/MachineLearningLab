```
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

# 显示中文(根据自己电脑设置)
plt.rcParams['font.sans-serif'] = ['SimHei']
# 显示负数
plt.rcParams['axes.unicode_minus'] = False

# 数据预处理
def pre_process(x, y):
    x -= np.mean(x, axis=0);
    x /= np.std(x, axis=0, ddof=1);
    x = np.c_[np.ones(len(x)), x];
    y = np.c_[y]
    print(x)
    return x, y

def model(x, theta):
    z = np.dot(x,theta)
    return 1/(1+np.exp(-z))

def cost_func(h, y):
    m = len(y)
    j = 1/m*(np.sum(-np.dot(y.T, np.log(h)) - np.dot((1 - y).T, np.log(1 - h))))
    return j

# gradient descent
def gra_desc(x, y, alpha, iter_num): 
    m, n = x.shape
    theta = np.zeros((n, 1))
    j_history = np.zeros(iter_num)
    
    for i in range(iter_num):
        h = model(x, theta);
        j_history[i] = cost_func(h, y);
        delta_theta = 1.0/m*np.dot(x.T, h-y)
        theta -= alpha * delta_theta;
    return j_history, theta

```

```
data = pd.read_csv('homework/ch.3/iris.csv', header=None)
data = data.values
x = data[:,:-1]
y = data[:,-1]

for i in range(len(y)):
    if y[i] == 'Iris-setosa':
        y[i] = 0;
    elif y[i] == 'Iris-versicolor':
        y[i] = 1;
    elif y[i] == 'Iris-virginica':
        y[i] = 2;

print(y[:10])
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=10)
```

```
clf_10 = GridSearchCV(estimator=LogisticRegression(), param_grid={'C': [1e-2, 1, 1e2]}, cv=10)
clf_10.fit(x_train, y_train)
best_model_10 = clf_10.best_estimator_

clf_1 = GridSearchCV(estimator=LogisticRegression(), param_grid={'C': [1e-2, 1, 1e2]}, cv=len(x_train))
clf_1.fit(x_train, y_train)
best_model_1 = clf_1.best_estimator_

acc_10 = best_model_10.score(x_test, y_test)
acc_1 = best_model_1.score(x_test, y_test)

print("10折交叉验证：", acc_10)
print("留一法：", acc_1)
```
