```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#读取并可视化训练集
# train_data = pd.read_csv("data/data.csv",names = ['A','B'])
# train_data.plot(kind = 'scatter',x = 'A',y = 'B', figsize = (12,8))
# plt.show()

train_data = np.genfromtxt('data/data.csv',delimiter=',')
x_data = train_data[:,0]
y_data = train_data[:,1]
```

```python
lr = 0.000001
theta0 = 0
theta1 = 0
iterations = 1000

#画图用
iterationArray = []
lostArray = []
#画图用
```

```python
#cost function
def costfunction(x,y,theta0,theta1):
    sum = 0
    for i in range(len(x)):
        sum+=np.power((theta0+theta1*x[i])-y[i],2)
    return sum/(2*len(x))
```

```python
def gradientDescent(x_data,y_data,theta0,theta1,iterations,lr):
    m = float(len(x_data))
    for j in range(iterations):
        sum1 = 0
        sum2 = 0
        for i in range(len(x_data)):
            sum1 += (theta0+theta1*x_data[i])-y_data[i]
            sum2 += ((theta0+theta1*x_data[i])-y_data[i])*x_data[i]
        theta0 = theta0 - lr*sum1/m
        theta1 = theta1 - lr*sum2/m
        lostAns = costfunction(x_data,y_data,theta0,theta1)

        iterationArray.append(j)
        lostArray.append(lostAns)
        
        print('迭代次数 = ',j,'损失 =',lostAns,'theat0 = ',theta0,'theat1 = ',theta1)
    return theta0,theta1
```

```python
theta0,theta1 = gradientDescent(x_data,y_data,theta0,theta1,iterations,lr)
```

```python
plt.plot(x_data,y_data,'b.')
plt.plot(x_data,theta0+theta1*x_data,'r')
plt.show()
```
